```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['category']=le.fit_transform(dataset['category'])
dataset['cuisine']=le.fit_transform(dataset['cuisine'])
```

```python
x=dataset.iloc[:,0:7].values
y=dataset.iloc[:,7:8].values
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder()
z=oh.fit_transform(x[:,1:2]).toarray()
q=oh.fit_transform(x[:,4:5]).toarray()
x=np.delete(x,1,axis=1)
x=np.delete(x,4,axis=1)
x=np.concatenate((q,z,x),axis=1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 0)
```

```python
x_train
```

```python
y_train
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc  = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(n_jobs = 1000)
regressor.fit(x_train, y_train)
```

```python
predict = regressor.predict(x_test)
```

```python
predict
```

```python
from sklearn.metrics import r2_score
acc1 = r2_score(y_test, predict)
acc1
```

```python
import pickle
pickle.dump(regressor,open('base_price.pkl','wb'))
```
